```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import RobustScaler,robust_scale
from sklearn.preprocessing import scale
```

```python
df = pd.read_csv('Datasets/diabetes.csv')
df.boxplot(column=['Age','Glucose','BMI'],figsize=(12,8))
plt.plot()
```

```python
outcome = df['Outcome']
df = df.drop(['Outcome'],axis=1)
```

```python
robust_scaled_df = pd.DataFrame(robust_scale(df),columns = df.columns)
```

```python
robust_scaled_df[['Glucose','BMI','Age']].describe()
robust_scaled_df.boxplot(column=['Glucose','BMI','Age'],figsize=(12,8))
plt.plot()
```

```python
robust_scaler = RobustScaler(copy=True,with_centering=True,with_scaling=True)
robust_scaler = robust_scaler.fit_transform(df)
robust_scaler_df = pd.DataFrame(robust_scaler,columns = df.columns)
robust_scaler_df.boxplot(column=['Glucose','BMI','Age'],figsize=(12,8))
plt.plot()
```
